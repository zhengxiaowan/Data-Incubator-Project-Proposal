```python
import pandas as pd
import numpy as np
import plotly 
plotly.tools.set_credentials_file(username='kelly312', api_key='arftQRcWSSkAySZ8uw18')
import plotly.plotly as py


df=pd.read_csv('h1b.csv')
df['STATE'] = df['WORKSITE'].apply(lambda x:x.split(',')[-1][1:])
df_certified=df[(df['CASE_STATUS']=='CERTIFIED')]
df_state=df_certified.groupby(['STATE']).agg({'PREVAILING_WAGE':['count','mean','std']})
df_state.reset_index(inplace=True)
df_state.columns= ['state','value','mean','std']
df_state['statecode']=df_state['state']
us_state_abbrev = {
    'ALABAMA': 'AL',
    'ALASKA': 'AK',
    'ARIZONA': 'AZ',
    'ARKANSAS': 'AR',
    'CALIFORNIA': 'CA',
    'COLORADO': 'CO',
    'CONNECTICUT': 'CT',
    'DELAWARE': 'DE',
    'FLORIDA': 'FL',
    'GEORGIA': 'GA',
    'HAWAII': 'HI',
    'IDAHO': 'ID',
    'ILLINOIS': 'IL',
    'INDIANA': 'IN',
    'IOWA': 'IA',
    'KANSAS': 'KS',
    'KENTUCKY': 'KY',
    'LOUISIANA': 'LA',
    'MAINE': 'ME',
    'MARYLAND': 'MD',
    'MASSACHUSETTS': 'MA',
    'MICHIGAN': 'MI',
    'MINNESOTA': 'MN',
    'MISSISSIPPI': 'MS',
    'MISSOURI': 'MO',
    'MONTANA': 'MT',
    'NEBRASKA': 'NE',
    'NEVADA': 'NV',
    'NEW HAMPSHIRE': 'NH',
    'NEW JERSEY': 'NJ',
    'NEW MEXICO': 'NM',
    'NEW YORK': 'NY',
    'NORTH CAROLINA': 'NC',
    'NORTH DAKOTA': 'ND',
    'OHIO': 'OH',
    'OKLAHOMA': 'OK',
    'OREGON': 'OR',
    'PENNSYLVANIA': 'PA',
    'RHODE ISLAND': 'RI',
    'SOUTH CAROLINA': 'SC',
    'SOUTH DAKOTA': 'SD',
    'TENNESSEE': 'TN',
    'TEXAS': 'TX',
    'UTAH': 'UT',
    'VERMONT': 'VT',
    'VIRGINIA': 'VA',
    'WASHINGTON': 'WA',
    'WEST VIRGINIA': 'WV',
    'WISCONSIN': 'WI',
    'WYOMING': 'WY',
    'DISTRICT OF COLUMBIA': 'DC',
    'PUERTO RICO': 'PR'
}
df_state_new=df_state.drop([27]).replace({'statecode': us_state_abbrev})

#plot 2

for col in df_state_new.columns:
    df_state_new[col] = df_state_new[col].astype(str)

scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]

df_state_new['text'] = df_state_new['state'] + '<br>' +\
    'Mean Wage $'+df_state_new['mean']+ '<br>' +\
    'Std Wage $'+df_state_new['std']

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = df_state_new['statecode'],
        z = df_state_new['value'].astype(float),
        locationmode = 'USA-states',
        text = df_state_new['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            )
        ),
        colorbar = dict(
            title = 'Petition Number'
        )
    ) ]

layout = dict(
        title = '2011-2016 H1B Petition Distribution',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)',
        ),
    )

fig = dict( data=data, layout=layout )
py.iplot( fig, filename='h1b-cloropleth-map' )
```
